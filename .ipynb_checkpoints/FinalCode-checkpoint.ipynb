```python
#(full/empty , high/low)
oven_status = [[False,False], [False,False], [False,False], [False,False], [False,False], [False,False]]

#append tuple (Time In, Pallet No., Duration)
oven_stack = [[],[],[],[],[],[]]

oven_time = {'A0':(320,43),
       'A1':(320,43),
       'A2':(450,43),
       'A3':(450,43),
       'A4':(460,43),
       'A5':(460,43),
       'A6':(460,43),
       'B2':(300,71),
       'B3':(300,71),
       'B4':(450,71),
       'B5':(450,71),
       'B6':(450,71),
       'C0':(240,43),
       'C1':(240,43),
       'C2':(270,43),
       'C3':(270,43),
       'C4':(300,43),
       'C5':(300,43),
       'C6':(300,43)}

for index, row in input_stack.iterrows():
    time_stamp = row.EndDate
    pallet_no = row.Product
    
    (duration,cooling_temp) = oven_time[pallet_no]
    old_duration = 0
    old_time_stamp = datetime.datetime(2022,2,1)
    
    oven_clocked = [len(oven_stack[0]), len(oven_stack[1]), len(oven_stack[2]), len(oven_stack[3]), len(oven_stack[4]), len(oven_stack[5])]
    oven_clocked = [x - min(oven_clocked) for x in oven_clocked]
    oven_pref_index = oven_clocked.index(0)
    print(oven_clocked)
    print(oven_pref_index)
    for i in range(len(oven_status)):
        
        if oven_status[oven_pref_index][0] == False:
            oven_status[oven_pref_index][0] = True
            oven_stack[oven_pref_index].append((time_stamp,pallet_no,duration))
            if cooling_temp == 43:
                oven_status[oven_pref_index][1] = False
            else:
                oven_status[oven_pref_index][1] = True
            break
            
        if oven_status[i][0] == False and oven_status[i][1] == True:
            oven_status[i][0] = True
            oven_stack[i].append((time_stamp,pallet_no,duration))
            if cooling_temp == 43:
                oven_status[i][1] = False
            break
            
        if oven_status[i][0] == False and oven_status[i][1] == False:
            oven_status[i][0] = True
            oven_stack[i].append((time_stamp,pallet_no,duration))
            if cooling_temp == 71:
                oven_status[i][1] = True
            break      
            
        if ((old_time_stamp-datetime.datetime(2022,2,1)).total_seconds() + old_duration) <= (time_stamp-datetime.datetime(2022,2,1)).total_seconds():        
            oven_status[i][0] = False

    old_time_stamp = time_stamp        
    old_duration = duration
```

```python
def to_visualize(oven_stack):
    
    results = []
    
    for oven_number in range(len(oven_stack)):
        default_time_stamp = oven_stack[oven_number][0][0]
        #print(type(default_time_stamp))
        
        for i in range(len(oven_stack[oven_number])):
            time_stamp = oven_stack[oven_number][i][0]
            product_name = oven_stack[oven_number][i][1]
            duration = oven_stack[oven_number][i][2]
            start_time = (time_stamp-default_time_stamp).total_seconds()
            finish_time = (start_time + duration*60)
        
            results.append({'Job': product_name,
                 'Machine': oven_number+1,
                 'Start': start_time,
                 'Duration': duration,
                 'Finish': finish_time})
            
            # results.append({'Task': product_name,
            #                 'Start': start_time,
            #                 'Finish':finish_time,
            #                 'Resource':oven_number+1})
                                   
    return results
```

```python
results= to_visualize(oven_stack)

df = pd.DataFrame(results, columns=['Job', 'Machine', 'Start', 'Duration', 'Finish'])
df=df.rename(columns={"Job": "Task"})
df.Start = pd.to_datetime(df.Start, unit='s', errors='coerce')
df.Finish = pd.to_datetime(df.Finish, unit='s', errors='coerce')
df.Duration = df.Duration * 60
df = df[['Machine','Start','Duration', 'Finish','Task']]
df['Machine'] = 'O' + df['Machine'].astype(str)
print(df)

colors = {1: 'rgb(220, 0, 0)',
          2: (1, 0.9, 0.16),
          3: 'rgb(0, 255, 100)',
          4: 'rgb(220, 0, 0)',
          5: (1, 0.9, 0.16),
          6: 'rgb(0, 255, 100)',}

fig = ff.create_gantt(df, group_tasks=True, show_colorbar=True)
#fig = ff.create_gantt(df, colors=colors, index_col='Machine', show_colorbar=True)
#fig.update_yaxes(automargin=True)
fig.show()
```
